```python
import pandas as pd
import pymongo

# Connect to MongoDB
client = pymongo.MongoClient("mongodb://192.168.3.82:27017/")

# List existing databases
print("Existing databases:")
print(client.list_database_names())
```

```python
# Create a new database (MongoDB creates it when you insert data)
# Just reference the database name - it will be created automatically

db_name = "vehicle_insurance"  # Change this to your desired database name
collection_name = "vehicle_insurance_data"  # Change this to your desired collection name

db = client[db_name]
collection = db[collection_name]

print(f"Database '{db_name}' and collection '{collection_name}' are ready!")
print(f"Note: The database will appear in the list only after inserting data.")
```

```python
# Load CSV data
df = pd.read_csv('data.csv')
print(f"Loaded {len(df)} records from CSV")
df.head()
```

```python
# Convert DataFrame to list of dictionaries and insert into MongoDB
data = df.to_dict(orient='records')

# Insert data into MongoDB (this also creates the database and collection)
collection.insert_many(data)
print(f"Successfully inserted {len(data)} records into MongoDB!")

# Verify the database now exists
print("\nDatabases after insertion:")
print(client.list_database_names())
```

```python
# Retrieve data from MongoDB
cursor = collection.find()
retrieved_data = list(cursor)

# Convert to DataFrame
df_from_mongo = pd.DataFrame(retrieved_data)

# Drop the MongoDB _id column
if '_id' in df_from_mongo.columns:
    df_from_mongo = df_from_mongo.drop('_id', axis=1)

print(f"Retrieved {len(df_from_mongo)} records from MongoDB")
df_from_mongo.head()
```
